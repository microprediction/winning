## Minimum of k random variables

See the [article](https://medium.com/@mike.roweprediger/density-of-the-minimum-of-n-random-variables-on-a-lattice-45583cc0c2d0).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skewnorm
import matplotlib as mpl

# Set plot size
mpl.rcParams['figure.figsize'] = [10.0, 6.0]


# Generate five densities
L = 50
x = np.linspace(-4, 4, 2*L+1)
density1 = norm.pdf(x, loc=-1, scale=0.5)
density2 = norm.pdf(x, loc=0, scale=0.8)
density3 = skewnorm.pdf(x, a=-3, loc=1, scale=1)
density4 = skewnorm.pdf(x, a=2, loc=-0.5, scale=0.5)

# Plot the densities
plt.plot(x, density1, label='Density 1')
plt.plot(x, density2, label='Density 2')
plt.plot(x, density3, label='Density 3')
plt.plot(x, density4, label='Density 4')
plt.legend()
plt.show()
```
